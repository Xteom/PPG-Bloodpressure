```python
# Read data
df = pd.read_csv('GTCARDIOAdvancedStudy.csv')
df
```

```python
# get the top values in heart rate (BPM) column 
top_heart_rate_vals = df['Heart Rate (BPM)'].nlargest(5)
top_heart_rate_vals
# as you can see top_heart_rate_vals is a list of the top 5 values

```

```python
# since it is a series, we can iterate through it
for val in top_heart_rate_vals:
    print(val)
```

```python
df.columns
```

```python
# get the activity type for the top 5 heart rates
for val in top_heart_rate_vals:
    print(df[df['Heart Rate (BPM)'] == val]['Activity (Level 1) '])
```

```python
# another way to do it

# now we'll do it using the index
for index in top_heart_rate_vals.index:
    print(df.loc[index]['Activity (Level 1) '])
```

The first way you are querying the data to get a dataframe where the value matches and the accessing to the column you want. The second way you are accessing directly to the row with the value you want (since nlargest() already gives you the index) and then to the column you want.

```python
# to get the values of an specific date, you first should filter the dataframe by the date
df_filtered = df[df['Date'] == '8/7'] 
df_filtered
```

```python
# now that we have the df filtered by the date, we can get the top 5 heart rates
top_heart_rate_vals_filtered = df_filtered['Heart Rate (BPM)'].nlargest(5)
for index in top_heart_rate_vals_filtered.index:
    print(df_filtered.loc[index]['Activity (Level 1) '])
```
